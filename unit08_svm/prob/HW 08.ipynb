```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
x = np.array([0, 1.3, 2.1, 2.8, 4.2, 5.7])
y = np.array([-1, -1, -1, 1,-1, 1])

J = np.vectorize (lambda t :
                  np.sum(
                        (1-y*(x-t))*
                        ((1-y*(x-t))>0).astype(int)
                        )
                 )
```

```python
t = np.arange(100)/20
Jt = J(t)
```

```python
plt.scatter(t,Jt)
print(t [np.argmin(Jt)], np.min(Jt))

(1-y*(x-3.1))*((1-y*(x-3.1))>0).astype(int)
```

```python
x = np.array([0,1,2,3])
y = np.array([1,-1,1,-1])
alpha = [0,0,1,1]
gamma = 3
```

```python
Z = np.vectorize (lambda k : np.sum(
                    alpha*y*np.exp(-gamma*((x-k)**2))
                ))
```

```python
k = np.arange(0,100)/20
plt.scatter(k, Z(k))
```

```python
plt.scatter(k, 2*(Z(k)>0).astype(int)-1)
```
